## Importing Libraries

```python
import cv2, numpy as np, matplotlib.pyplot as plt
```

## Orignal Image

```python
image1 = cv2.imread("images/star0.jpg")
plt.figure(figsize=(15,20))
plt.title("Colored image")
plt.imshow(image1)
```

## Converting Orignal image to grayscale

```python
gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15,20))
plt.title("Grayscale image")
plt.imshow(gray)
```

## Brighten up Image

Each pixel in these images are between 0 and 255 denoting their brightness.
Here stars are brighter than the black (actually not so black) background.
We will look if any pixel is having value greater than threshold, if it does than we will make that value 255 (Maximum brightness) and if it doesn't than we will make it 0.

```python
gray2 = np.where(gray >= 60, 255, gray)
gray2 = np.where(gray < 60, 0, gray)
plt.figure(figsize=(15,20))
plt.title("Grayscale image Brighten Up")
plt.imshow(gray2)
```

# Above operations on bulk images

### Converting colored images into grayscale

```python
n = int(input("Enter number of images:"))
for i in range(n):
    img = cv2.imread("images/star{}.jpg".format(i))
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite("images/gray-star{}.png".format(i), gray_img)
```

### Brighten up stars

```python
n = int(input("Enter number of images:"))
th = int(input("Enter pixel threshold:"))
for i in range(n):
    gray = cv2.imread("images/gray-star{}.png".format(i))
    gray2 = np.where(gray >= th, 255, gray)
    gray2 = np.where(gray < th, 0, gray)
    cv2.imwrite("images/gray-star{}.png".format(i), gray2)
```

## Stacking images

Images will be stacked on one another to reduce singnal-to-noise ratio.\
Unfortunately, my images are a bit offset so I only stacked 2 images which were having lowest offset.

```python
img1 = cv2.imread("images/gray-star6.png")
img2 = cv2.imread("images/gray-star7.png")
added_image = cv2.addWeighted(img1,0.5,img2,0.5,0)
cv2.imwrite("images/final.png", added_image)
```
